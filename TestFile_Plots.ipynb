```python
def mkdir_p(mypath):
    '''Creates a directory. equivalent to using mkdir -p on the command line'''

    from errno import EEXIST
    from os import makedirs,path

    try:
        makedirs(mypath)
    except OSError as exc: # Python >2.5
        if exc.errno == EEXIST and path.isdir(mypath):
            pass
        else: raise
```

```python
def DoesDirectoryExist(mypath): #extra precaution (Probably overkill...)
    '''Checks to see if Directory exists before running mkdir_p'''
    import os.path
    from os import path
    
    if path.exists(mypath):
        pass
    else:
        mkdir_p(mypath)
```

# "Probet" = 
#### | anti-tagged jet > < t-tagged probe jet |
# "at" =
#### | anti-tagged jet > < any probe jet |
# "pret" = 
#### | t-tagged jet > < any jet |
# "0t" = 
#### | no tag jet > < no tag jet |
# "1t" = 
#### | t-tagged jet > < no tag jet |  or  | no tag jet > < t-tagged jet |
# "1t+2t" = 
#### | t-tagged jet > < no tag jet |  or  | no tag jet > < t-tagged jet |  or  | t-tagged jet > < t-tagged jet |
# "2t" = 
#### | t-tagged jet > < t-tagged jet |
# "0t+1t+2t" = 
#### | any jet > < any jet | 

```python
import matplotlib.colors as colors
import matplotlib.pyplot as plt
import warnings
import re # regular expressions
warnings.filterwarnings("ignore")

# ---- Reiterate categories ---- #
ttagcats = ["Probet", "at", "0t", "pret", "1t", "1t+2t", "2t", "0t+1t+2t"] # 'Probet' = t-tagged probe and anti-tag pair
btagcats = ["0b", "1b", "2b"]
ycats = ['cen', 'fwd']

list_of_cats = [ t+b+y for t,b,y in itertools.product( ttagcats, btagcats, ycats) ]

# ---- List the Histograms Here ---- #
list_of_hists_4vector = ('SDmass', 'tau32', 'jetmass', 'ttbarmass', 'jetpt', 'jeteta', 'jetphi', 'jety', 'jetdy')
```

```python
Lum2016 = 35920.

ttbar_BR = 0.45 #Calculated using PDG 2019
ttbar_xs = 1.0#750.5 # from https://cms-gen-dev.cern.ch/xsdb/?columns=67108863&currentPage=0&pageSize=0&searchQuery=process_name%3DTTJets_TuneCP5%2A
# xs should've already been included in evt weights
ttbar_700to1000_xs = 1.0 # N/A on XSDB yet
ttbar_1000toInf_xs = 1.0#16.42 # \pm 0.05547 from XSDB
ttbar_total_xs = 1.0#831.76 # from https://twiki.cern.ch/twiki/bin/view/LHCPhysics/TtbarNNLO#Top_quark_pair_cross_sections_at

ttbar_sf = ttbar_xs*ttbar_BR*Lum2016/TTbar_unweighted['cutflow']['sumw']
ttbar_700to1000_sf = ttbar_700to1000_xs*ttbar_BR*Lum2016/TTbar_biased_700to1000_unweighted['cutflow']['sumw']
ttbar_1000toInf_sf = ttbar_1000toInf_xs*ttbar_BR*Lum2016/TTbar_biased_1000toInf_unweighted['cutflow']['sumw']
ttbar_total_sf = ttbar_total_xs*ttbar_BR*Lum2016/(TTbar_biased_700to1000_unweighted['cutflow']['sumw']+TTbar_biased_1000toInf_unweighted['cutflow']['sumw'])
# print(ttbar_1000toInf_sf)
```

```python
# stack_tag1_opts = {'marker': '.', 'markersize': 10., 'color': 'b', 'elinewidth': 1}
# stack_tag2_opts = {'marker': 's', 'markersize': 5., 'color': 'g', 'elinewidth': 1}
# legend_tag_opts = {'labels':['', '', 'DeepTag', 'DeepTag MD']}
# stack_ttbar_opts = {'alpha': 0.8, 'edgecolor':(0,0,0,0.3), 'color': 'red'}
stack_ttbar_opts = {'color': 'red'}
stack_ttbar_biased_opts = {'color': 'blue'}
line_ttbar_opts = {'color': 'r'}
line_ttbar_biased_opts1 = {'color': 'g'}
line_ttbar_biased_opts2 = {'color': 'b'}
# stack_rsg1000_opts = {'color': 'orange'}
# stack_background_opts = {'alpha': 0.8, 'edgecolor':(0,0,0,0.3), 'color': 'yellow'}
stack_error_opts = {'label':'Stat. Unc.', 'hatch':'///', 'facecolor':'None', 'edgecolor':(0,0,0,.5), 'linewidth': 0}
# data_err_opts = {'linestyle': 'none', 'marker': '.', 'markersize': 10., 'color': 'k', 'elinewidth': 1}
```

# Plot both 'old' $t\bar{t}$ samples and sum of biased samples together
## Normalize each separate histogram to 1

```python
for ihist in list_of_hists_4vector:
    SaveDirectory = maindirectory + '/ttbar_normal_and_biased_Plots/SumBiasedSamples/NormalizedToOne/LogScale/' + ihist + '/' # split histograms into subdirectories
    DoesDirectoryExist(SaveDirectory) # no need to create the directory several times if it exists already
    for icat in list_of_cats: 
        plt.rcParams.update({
        'font.size': 14,
        'axes.titlesize': 18,
        'axes.labelsize': 18,
        'xtick.labelsize': 12,
        'ytick.labelsize': 12
        })
        fig, (ax) = plt.subplots(
#             nrows=1,
#             ncols=1,
#             figsize=(7,7),
#             gridspec_kw={"height_ratios": (3, 1)},
#             sharex=True
        )
        fig.subplots_adjust(hspace=.07)

        title = ihist + '  ' + icat
        filename = ihist + '_' + icat + '.' + 'png'

        legend_labels = {'labels':[r'TTJets', r'TT_Mtt-700to1000 + TT_Mtt-1000toInf'],
                     'loc': 'upper right',
                     'fontsize': 'xx-small'}

        ttbar_hist = TTbar_unweighted[ihist].integrate('anacat', icat).integrate('dataset', 'TTbar')
#         ttbar_hist.scale(ttbar_sf)
        
        ttbar_biased_700to1000_hist = TTbar_biased_700to1000_unweighted[ihist].integrate('anacat', icat).integrate('dataset', 'TTbar_biased_700to1000')
#         ttbar_biased_700to1000_hist.scale(ttbar_700to1000_sf)
        
        ttbar_biased_1000toInf_hist = TTbar_biased_1000toInf_unweighted[ihist].integrate('anacat', icat).integrate('dataset', 'TTbar_biased_1000toInf')
#         ttbar_biased_1000toInf_hist.scale(ttbar_1000toInf_sf)
        
        if ihist == 'SDmass':
            Ntt = np.sum(ttbar_hist.integrate('jetmass').values()) # extract number of events from histogram directly
            Nttbiased1 = np.sum(ttbar_biased_700to1000_hist.integrate('jetmass').values())
            Nttbiased2 = np.sum(ttbar_biased_1000toInf_hist.integrate('jetmass').values())
        else:
            Ntt = np.sum(ttbar_hist.integrate(ihist).values())
            Nttbiased1 = np.sum(ttbar_biased_700to1000_hist.integrate(ihist).values())
            Nttbiased2 = np.sum(ttbar_biased_1000toInf_hist.integrate(ihist).values())
        Ntt = [i for i in Ntt.values()] # the zeroth column entry is the number of events defined as a float
        Nttbiased1 = [i for i in Nttbiased1.values()]
        Nttbiased2 = [i for i in Nttbiased2.values()]
        
        ttbar_hist.scale(1.0/Ntt[0])
#         ttbar_biased_700to1000_hist.scale(1.0/Nttbiased1[0])
#         ttbar_biased_1000toInf_hist.scale(1.0/Nttbiased2[0])
    
        ttbar_biased_sum = ttbar_biased_700to1000_hist.copy()
        ttbar_biased_sum.add(ttbar_biased_1000toInf_hist)
        ttbar_biased_sum.scale( 1.0/(Nttbiased1[0]+Nttbiased2[0]) )

        hist.plot1d(ttbar_hist, clear=False,
                    line_opts=line_ttbar_opts, error_opts=stack_error_opts)
        hist.plot1d(ttbar_biased_sum, clear=False,
                    line_opts=line_ttbar_biased_opts2, error_opts=stack_error_opts, legend_opts=legend_labels)

        ax.set_ylim(bottom=0.)
        ax.set_yscale('log')
        ax.set_ylabel('Events')
#         ax.set_xlabel(None)
        ax.set_title(title)
#         leg = ax.legend(labels=[r'TTJets', r'TT_Mtt-700to1000 + TT_Mtt-1000toInf'])

        if ihist == 'ttbarmass':
            ax.set_xlim(700,5000)
#             ax.set_xlim(100,5000)
        if ihist == 'jetpt':
            ax.set_xlim(400,3000)
        if ihist == 'jeteta':
            ax.set_xlim(-3,3)
        if ihist == 'tau32':
            ax.set_xlim(0,1.2)
            
        plt.autoscale(enable=True, axis='y', tight=True)
        plt.savefig(SaveDirectory+filename, bbox_inches="tight")
        print(SaveDirectory+filename)
```
